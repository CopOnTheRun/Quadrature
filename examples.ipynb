```python
from Interval import Interval, Method
from Quadrature import Riemann
%matplotlib notebook
```

```python
#the function to be graphed
function = lambda x: x**3

#create the interval
interval = Interval(-2,2)

#divide the interval into 10 partitions
interval //= 8

#choose the method
method = Method.mid()

#create the Riemann function
riemann = Riemann(function,interval,method)
```

```python
#calculate the sum
#note is should be ~0 because this is an odd function about an symettric interval
print(riemann.calc())

#graph the Riemann function
riemann.graph()
```

```python
#change the method used to grab each interval point
riemann.method = Method.left()
print(riemann.calc())
riemann.graph()
```

```python
#the same as above, but use the max point method
riemann.method = Method.max()
print(riemann.calc())
riemann.graph()
```

```python
#subdivide the interval into 16 parts instead of 8
riemann.interval //= 16

#notice how the calculated value is closer to the actual value
print(riemann.calc())
riemann.graph()
```

```python
#change the function being calculated
riemann.func = lambda x: x**2
print(riemann.calc())
riemann.graph()
```
